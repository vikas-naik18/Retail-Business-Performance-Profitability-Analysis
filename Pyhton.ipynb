```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('retail_data.csv')
```

```python
df['Net_Revenue']=df['TotalSales'] - df['Discount']
```

```python
correlation = df['Discount'].corr(df['TotalSales'])
print("Correlation:",correlation)
```

```python
sns.scatterplot(x='Discount',y='TotalSales',data=df)
plt.title('Discount vs Sales')
plt.show()
```
